```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('depressive_unigram_tweets_final.csv')
df
```

```python
import nltk
nltk.download(['punkt','stopwords'])
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
df.isnull().any().any()
```

```python
df.info(null_counts=True)
```

```python
df_nonull = df[df['tweet'].notnull()]
```

```python
df_nonull.info(null_counts = True)
```

```python
df_nonull['vader_score'] = df_nonull['tweet'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
```

```python
df_nonull.head()
```

```python
positive_num = len(df_nonull[df_nonull['vader_score'] >=0.05])
negative_num = len(df_nonull[df_nonull['vader_score']<0.05])
positive_num, negative_num
```

```python
df_nonull['vader_sentiment_label']= df_nonull['vader_score'].map(lambda x:int(1) if x>=0.05 else int(0))
df_nonull.head()
```

```python
group = df_nonull.groupby(df_nonull.vader_sentiment_label)
df_depress = group.get_group(0)
```

```python
df_depress.drop(['Unnamed: 0','Unnamed: 0.1','cashtags','hashtags','id','time'],axis = 1, inplace=True)
```
